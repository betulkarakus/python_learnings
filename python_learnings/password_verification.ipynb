```python
# Set the password
password = "123.qaz"

# Get the password from the user
user_password = input('Enter your password: ')

# Check the entered password.
# If incorrect, keep prompting until the user enters the correct password
while user_password != password:
    print("Incorrect password.")
    print("Please check your password.")

    # Get the password again from the user
    user_password = input('Enter your password: ')

# Print a congratulatory message if the password is correct
print("Congratulations, you have successfully logged in!")
```
