```python
# Creating Kitchen and Bathroom Shopping Lists
kitchen_list = ["Tomato", "Cheese", "Olives"]
bathroom_list = ["Soap", "Bleach"]

# Check if an item exists in the kitchen list.
item_to_check = input("Enter the item you want to check in your kitchen list: ")
if item_to_check in kitchen_list:
    print("{} is in your kitchen shopping list.".format(item_to_check))
else:
    print("{} is not in your kitchen shopping list.".format(item_to_check))

# Combine the kitchen and bathroom shopping lists into all_shopping_list using extend().
all_shopping_list = kitchen_list
all_shopping_list.extend(bathroom_list)
print("Combined shopping list for kitchen and bathroom:", all_shopping_list)
```
