```python
# Get the control value from the user (between 1 and 10)
control_value = int(input("Enter the control value (1..10): "))

# Iterate through numbers from 0 to 10 (exclusive)
for number in range(11):
    # Check if the current number is equal to the control value
    if number == control_value:
        # If equal, print a message and move to the next iteration using 'continue'
        print("Moving to the next iteration of the loop...")
        continue
        # The following statement is never executed because of the 'continue' above
        print("This statement is never executed")
    else:
        # Print the current number if it's not equal to the control value
        print(number)

# Print the first statement after the loop
print("First statement after the loop")
```
