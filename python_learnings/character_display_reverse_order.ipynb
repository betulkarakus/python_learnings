
# Description:
Character Display in Reverse Order using a For Loop.
This Python code prompts the user to enter a name and displays each character in reverse order using a for loop.


```python
# Prompt the user to enter a name
name = input("Enter a name: ")

# Iterate through each character in reverse order using a for loop
for character_index in range(len(name) - 1, -1, -1):  # Start from the last character's index to the first (in reverse order)
    # Retrieve the character at the current index
    character = name[character_index]

    # Print the current character
    print(character)
```
