
# Description:

This program defines a function sum_of_natural_numbers that takes a positive integer limit as input and calculates the sum of natural numbers up to that limit using the sum function and the range function. The if __name__ == "__main__": block prompts the user to enter a limit and then calculates and prints the sum of natural numbers up to that limit.


```python
def sum_of_natural_numbers(limit):
    # Ensure the limit is a positive integer
    if limit <= 0 or not isinstance(limit, int):
        print("Please enter a positive integer as the limit.")
        return

    # Calculate the sum of natural numbers
    sum_result = sum(range(1, limit + 1))

    print(f"The sum of natural numbers up to {limit} is: {sum_result}")

if __name__ == "__main__":
    # Get user input for the limit
    limit = int(input("Enter the limit for the sum of natural numbers: "))

    # Calculate and display the sum of natural numbers
    sum_of_natural_numbers(limit)
```
