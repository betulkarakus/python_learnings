
# Description:
This program generates a set of 6 random lottery numbers between 1 and 49 without repetition. The numbers are then sorted and printed to simulate a lottery ticket.


```python
# Import the random module
import random

# Create an empty list to store the lottery numbers
lottery_numbers = []

# Generate 6 random numbers for the lottery ticket
for i in range(6):
    # Generate a random number between 1 and 49
    random_number = random.randint(1, 49)

    # Ensure the generated number is not repeated
    # Use a while loop to generate a new number as long as it's already in the list
    while random_number in lottery_numbers:
        random_number = random.randint(1, 49)

    # Add the random_number to the list of lottery_numbers
    lottery_numbers.append(random_number)

# Sort the generated lottery numbers
lottery_numbers.sort()

# Print the generated lottery numbers
print('Your lottery numbers:')
for number in lottery_numbers:
    print(number)

print("Good luck, but don't leave it all to chance :)")
```
