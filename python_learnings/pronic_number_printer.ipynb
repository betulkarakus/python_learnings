
# Description:
This Python program prints all pronic numbers between 1 and 100.

A pronic number (or rectangular number, or oblong number) is a number that is the product of two consecutive integers. It can be represented as n * (n + 1), where n is an integer.



```python
def is_pronic_number(number):
    """
    Checks if the given number is a pronic number.

    Parameters:
    - number (int): The input number to be checked.

    Returns:
    - bool: True if the number is a pronic number, False otherwise.
    """
    for n in range(int(number**0.5) + 1):
        if n * (n + 1) == number:
            return True
    return False

if __name__ == "__main__":
    print("Pronic numbers between 1 and 100:")
    for i in range(1, 101):
        if is_pronic_number(i):
            print(i)
```
