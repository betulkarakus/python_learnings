```python
# 1. Get user inputs
fixed_fee = float(input('Enter your monthly fixed internet fee: '))
extra_used_amount = int(input('Enter the additional internet usage amount in GB: '))
category = int(input('Enter your internet category number: '))

# 2. Calculate the total bill based on the category

# Check if the category is within valid range (1 to 3)
if 1 <= category <= 3:
    # Calculate the total bill based on the specified rules for each category
    if category == 1:
        total_bill = fixed_fee + extra_used_amount * 0.10
    elif category == 2:
        total_bill = fixed_fee + extra_used_amount * 0.20
    else:
        total_bill = fixed_fee + extra_used_amount * 0.30
    # Display the total bill amount
    print('Your total bill amount is: ', total_bill)
else:
    # Display an error message if the category is outside the valid range
    print('You entered an incorrect category! Please check your category and try again.')

# Display the end of the program
print('End of the program')
```
