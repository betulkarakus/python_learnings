
# Description:
This program defines a function to perform addition, subtraction, multiplication, and division on two numbers entered by the user. The results are then displayed with informative messages.


```python
# Function definition to perform four arithmetic operations on two numbers
def four_operations(number1, number2):
    result_addition = number1 + number2
    result_subtraction = number1 - number2
    result_multiplication = number1 * number2
    result_division = number1 / number2

    # Return the four values
    return result_addition, result_subtraction, result_multiplication, result_division

# Main program

# Get the numbers from the user
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))

# Call the function to perform four operations
sum_result, difference_result, product_result, quotient_result = four_operations(number1, number2)

# Display the results
print("Sum: ", sum_result)
print("Difference: ", difference_result)
print("Product: ", product_result)
print("Quotient: ", quotient_result)
```
