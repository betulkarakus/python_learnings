#Problem:

A triangle can be classified as equilateral, isosceles, or scalene based on the lengths of its sides. An equilateral triangle has all three sides of equal length. An isosceles triangle has two sides of equal length and a third side of a different length. If all side lengths are different, the triangle is scalene. Write a program that reads the lengths of the three sides of a triangle from the user. Display a message indicating the type of the triangle.

###Note:
Before determining the type of the triangle, first check whether a triangle can be formed with the given side lengths by testing the "Triangle Inequality Theorem."

```python
# Take the lengths of the sides of a triangle from the user
s1 = float(input("Please enter the length of the first side of a triangle: "))
s2 = float(input("Please enter the length of the second side of a triangle: "))
s3 = float(input("Please enter the length of the third side of a triangle: "))

# Test the Triangle Inequality Theorem
# Can a triangle be formed with the given side lengths?
a = s1 - s2
b = s1 - s3
c = s2 - s3

if (s1 + s2) > s3 > abs(a) and (s1 + s3) > s2 > abs(b) and (s2 + s3) > s1 > abs(c):
    triangle_exists = True
else:
    triangle_exists = False

# If a triangle can be formed with the given sides
# Determine the type of the triangle
if triangle_exists:
    if s1 == s2 == s3:
        triangle_type = "equilateral"
    elif s1 == s2 or s1 == s3 or s2 == s3:
        triangle_type = "isosceles"
    else:
        triangle_type = "scalene"

    # Display the type of the triangle
    print("This is a {} triangle.".format(triangle_type))
else:
    # A triangle cannot be formed with the given side lengths!
    print("A triangle cannot be formed.")
```
