
# Description:
This Python program demonstrates how to insert a key-value pair at the beginning of an OrderedDict.

```python
from collections import OrderedDict

# Function to insert a key-value pair at the beginning of an OrderedDict
def insert_at_beginning(ordered_dict, key, value):
    """
    Insert a key-value pair at the beginning of an OrderedDict.

    Parameters:
    - ordered_dict: The OrderedDict to which the key-value pair will be inserted.
    - key: The key to be inserted.
    - value: The value corresponding to the key.

    Returns:
    - Updated OrderedDict.
    """
    # Create a new OrderedDict with the desired key-value pair
    new_dict = OrderedDict([(key, value)])

    # Update the new_dict with the original ordered_dict
    new_dict.update(ordered_dict)

    return new_dict

# Example OrderedDict
ordered_dict = OrderedDict([('a', 1), ('b', 2), ('c', 3)])

# Key and value to insert at the beginning
new_key = 'z'
new_value = 0

# Insert at the beginning
updated_ordered_dict = insert_at_beginning(ordered_dict, new_key, new_value)

# Display the updated OrderedDict
print("Original OrderedDict:", ordered_dict)
print("Updated OrderedDict with insertion at the beginning:", updated_ordered_dict)
```
