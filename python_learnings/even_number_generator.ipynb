
# Description:
This program takes an input 'n' from the user and generates a list of even numbers between 0 and n.
The even numbers are then printed in a comma-separated form.



```python
def even_numbers_generator(n):
    """Generator function to yield even numbers between 0 and n."""
    for i in range(0, n+1, 2):
        yield i

def main():
    try:
        # Get user input for n
        n = int(input("Enter a value for n: "))

        # Validate if n is a non-negative integer
        if n < 0:
            print("Please enter a non-negative integer.")
            return

        # Generate even numbers using the generator
        even_numbers = even_numbers_generator(n)

        # Join the even numbers and print in comma-separated form
        result = ",".join(map(str, even_numbers))
        print(f"Even numbers between 0 and {n}: {result}")

    except ValueError:
        print("Please enter a valid integer.")

if __name__ == "__main__":
    main()
```
