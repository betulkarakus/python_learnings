# Description:
This Python code takes a user-inputted name and displays each character one by one using a while loop.


```python
# Prompt the user to enter a name
name = input("Enter a name: ")

# Initialize the character index to 0
character_index = 0

# Iterate through each character in the name using a while loop
while character_index < len(name):
    # Retrieve the character at the current index
    character = name[character_index]

    # Print the current character
    print(character)

    # Increment the character index to move to the next character
    character_index += 1
```
