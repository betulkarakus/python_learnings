# Description:
This Python code defines a function 'kelime_varmi' that checks if a given word exists in a given text.
The user is prompted to enter a text and a word for searching, and the function determines whether the word is present in the text.


```python
# Function to check if a word exists in a text
def word_exists(searched_word, txt):
    # Convert both the searched word and text to lowercase for case-insensitive comparison
    searched_word = searched_word.lower()
    txt = txt.lower()

    # Check if the word is present in the text
    if searched_word in txt:
        print("'{}' word is present in the expression '{}'.".format(searched_word, txt))
    else:
        print("'{}' word is not present in the expression '{}'.".format(searched_word, txt))

# Main Program
# Prompt the user to enter the text for searching
text_to_search = input("Enter the text for searching: ")

# Prompt the user to enter the word to search for
word_to_search = input("Enter the word to search for: ")

# Call the word search function
word_exists(word_to_search, text_to_search)
```
