
# Description:
This program calculates the total invoice amount for Internet usage based on the monthly fixed fee, extra data used, and the category of the Internet plan for multiple customers.


```python
# Function to calculate the total invoice amount for Internet usage
def calculate_invoice():
    # Get user input for fixed fee, extra data used, and Internet plan category
    fixed_fee = float(input('Enter your monthly fixed Internet fee: '))
    extra_used_data = int(input('Enter the extra data used in GB: '))
    category = int(input('Enter your Internet plan category number: '))

    # Check if the category is within the valid range
    if 1 <= category <= 3:
        # Calculate the total invoice amount based on the category
        if category == 1:
            total_invoice = fixed_fee + extra_used_data * 0.10
        elif category == 2:
            total_invoice = fixed_fee + extra_used_data * 0.20
        else:
            total_invoice = fixed_fee + extra_used_data * 0.30
        return total_invoice
    else:
        # Return None for invalid category
        return None

# Get the number of customers from the user
customer_count = int(input("Enter the number of customers: "))

# Loop through each customer and calculate their invoice
for customer in range(1, customer_count + 1):
    print("For customer {}: ".format(customer))
    invoice = calculate_invoice()

    # Display the result or an error message if the invoice could not be calculated
    if invoice is not None:
        print('Your total invoice amount is: ', invoice)
    else:
        print('Your invoice could not be calculated due to incorrect category input.')

print('End of the program')
```
