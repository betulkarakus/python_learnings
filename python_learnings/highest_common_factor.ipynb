
# Description:
In this program, the find_hcf function calculates the Highest Common Factor (HCF) using the Euclidean algorithm. The while y loop continues until y becomes zero, and in each iteration, the values of x and y are updated accordingly. The result is the HCF of the two input numbers.

The if __name__ == "__main__": block prompts the user to enter two numbers and then calculates and prints the HCF of those two numbers.

Here's a Python program to find the Highest Common Factor (HCF) of two numbers:

```python
def find_hcf(x, y):
    while y:
        x, y = y, x % y
    return x

if __name__ == "__main__":
    # Get user input for two numbers
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))

    # Calculate and display the HCF
    hcf_result = find_hcf(num1, num2)
    print(f"The Highest Common Factor (HCF) of {num1} and {num2} is: {hcf_result}")
```
