```python
# 1. Get user inputs
temperature = int(input("Enter the temperature in Celsius: "))

# 2. Determine the state of matter and display the result

# Check if the temperature is above freezing point
if temperature > 0:
    # If temperature is between 0 and 100, the state of matter is 'LIQUID'
    if 0 < temperature < 100:
        state_of_matter = 'LIQUID'
    # If temperature is above 100, the state of matter is 'GAS'
    else:
        state_of_matter = 'GAS'
# If temperature is at or below freezing point, the state of matter is 'SOLID'
else:
    state_of_matter = 'SOLID'

# Display the result based on the determined state of matter
print("The substance is in {} state since the temperature is {} degrees Celsius.".format(state_of_matter, temperature))
print("End of the program")
```
