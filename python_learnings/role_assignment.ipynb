```python
# User role assignment based on the input code
# Get the user input for the access code
code = int(input("Enter your access code: "))

# Determine the user role based on the entered code
if code == 1:
    role = "Admin"
elif code == 2:
    role = "Editor"
elif code == 3:
    role = "Guest"
else:
    role = "Incorrect input! Please check your code."

# Display the assigned user role
print(role)
```
