
# Description:
This Python program checks if a given number is a Disarium number.

A **Disarium number** is a number defined by the following process: sum of its digits each raised to the power of its respective position is equal to the number itself. For example, 89 is a Disarium number because 8^1 + 9^2 = 8 + 81 = 89.



```python
def is_disarium_number(number):
    """
    Checks if the given number is a Disarium number.

    Parameters:
    - number (int): The input number to be checked.

    Returns:
    - bool: True if the number is a Disarium number, False otherwise.
    """
    str_number = str(number)
    length = len(str_number)
    digit_sum = sum(int(digit) ** (i + 1) for i, digit in enumerate(str_number))

    return digit_sum == number

if __name__ == "__main__":
    # Get user input for the number
    input_number = int(input("Enter a number: "))

    # Check if the number is a Disarium number
    result = is_disarium_number(input_number)

    # Display the result
    if result:
        print(f"{input_number} is a Disarium number.")
    else:
        print(f"{input_number} is not a Disarium number.")
```
