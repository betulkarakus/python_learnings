
# Description:
This Python program calculates the number of letters and digits in a given sentence.



```python
def count_letters_and_digits(sentence):
    """
    Count the number of letters and digits in a sentence.

    Parameters:
    - sentence: The input sentence.

    Returns:
    - Tuple containing the count of letters and digits.
    """
    letters_count = 0
    digits_count = 0

    # Iterate through each character in the sentence
    for char in sentence:
        if char.isalpha():
            # Increment the letters count if the character is a letter
            letters_count += 1
        elif char.isdigit():
            # Increment the digits count if the character is a digit
            digits_count += 1

    return letters_count, digits_count

# Accept a sentence from the user
user_sentence = input("Enter a sentence: ")

# Calculate the counts
letters_count, digits_count = count_letters_and_digits(user_sentence)

# Display the results
print("LETTERS", letters_count)
print("DIGITS", digits_count)
```
