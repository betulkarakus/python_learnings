
# Description:
This program defines a function to determine the largest among three numbers and then uses it to find the maximum number among user-input values.


```python
# Function to determine the largest of three numbers
def find_largest(a, b, c):
    # Compare the numbers to find the largest
    if a > b and a > c:
        largest = a
    elif b > a and b > c:
        largest = b
    elif c > a and c > b:
        largest = c
    else:
        # If there are at least two equal numbers, return None
        largest = None
    return largest

# Main program function
number_1 = int(input("Enter the first number: "))
number_2 = int(input("Enter the second number: "))
number_3 = int(input("Enter the third number: "))

result = find_largest(number_1, number_2, number_3)

# Display the result or a message if the largest number cannot be determined
if result is None:
    print("There is equality between at least two numbers! The largest number cannot be determined.")
else:
    print("The largest number among {}, {}, and {} is: ".format(number_1, number_2, number_3), result)
```
