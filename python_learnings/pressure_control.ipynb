
# Description:
This program calculates the pressure levels in a given number of containers based on user-provided parameters and checks if the pressure is within normal or dangerous ranges.


```python
def pressure_control(P):
    # Check if the pressure is below the dangerous level
    if P < dangerous_level:
        pressure_status = "NORMAL"
    else:
        pressure_status = "DANGEROUS"
    return pressure_status

R = 0.08206
container_count = int(input("Enter the number of containers: "))

for container in range(1, container_count + 1):
    print("For container {}: ".format(container))
    V = float(input("Enter the volume of the container: "))
    n = float(input("Enter the number of moles of the substance: "))
    T = int(input("Enter the temperature of the substance: "))
    dangerous_level = float(input("Enter the dangerous pressure level for this substance: "))

    # Calculate the pressure
    P = (n * R * T) / V

    # Check the pressure status using the function
    pressure_status = pressure_control(P)
    print("The pressure level in container {} is {}.".format(container, pressure_status))

print("End of the program")
```
