
# Description:
This Python program takes a sentence as input, splits it into words, and sorts the words in alphabetical order.



```python
def sort_words_alphabetically(sentence):
    """
    Sorts the words in a given sentence in alphabetical order.

    Parameters:
    - sentence (str): The input sentence containing words.

    Returns:
    - list: A list of words sorted in alphabetical order.
    """
    words = sentence.split()
    words.sort()
    return words

if __name__ == "__main__":
    # Get user input for the sentence
    input_sentence = input("Enter a sentence: ")

    # Sort the words alphabetically
    sorted_words = sort_words_alphabetically(input_sentence)

    # Display the sorted words
    print("\nWords in alphabetical order:")
    for word in sorted_words:
        print(word)
```
