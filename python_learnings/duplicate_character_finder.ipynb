
# Description:
This Python program finds all duplicate characters in a given string.



```python
def find_duplicate_characters(input_string):
    """
    Finds all duplicate characters in the given string.

    Parameters:
    - input_string (str): The input string to be checked.

    Returns:
    - list: A list of duplicate characters found in the string.
    """
    char_count = {}
    duplicates = []

    for char in input_string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    for char, count in char_count.items():
        if count > 1:
            duplicates.append(char)

    return duplicates

if __name__ == "__main__":
    # Get user input for the string
    input_string = input("Enter a string: ")

    # Find duplicate characters
    result = find_duplicate_characters(input_string)

    # Display the result
    if result:
        print("Duplicate characters in the string:", ", ".join(result))
    else:
        print("No duplicate characters found in the string.")
```
