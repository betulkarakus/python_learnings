```python
import random

def generate_name():
    # List of tuples containing first and last names
    full_names = [
        ("Albert", "Einstein"),
        ("Charles", "Darwin"),
        ("Nicolas", "Copernicus"),
        ("Michael", "Faraday"),
        ("Anders", "Celsius"),
        ("Isaac", "Newton"),
        ("Stephen", "Hawking"),
        ("Marie", "Curie"),
        ("Richard", "Dawkins")
    ]

    # Randomly select a tuple from the list
    first_name, last_name = random.choice(full_names)

    # Combine first and last names into a full name string
    return "{} {}".format(first_name, last_name)

# Generate and print five random names
for i in range(5):
    print(generate_name())
```
