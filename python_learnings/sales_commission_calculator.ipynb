#Problem:

According to the management policy of a company, a salesperson is paid a fixed salary along with a commission based on the total earnings from sales and a bonus rate. If the total earnings exceed the performance level defined by the company policy, the bonus rate is set to 0.10; otherwise, it is set to 0.05. Develop a program to calculate the salary of a salesperson in the company.



```python
# Define Variables & Take User Inputs
fixed_salary = float(input("Enter Fixed Salary: "))
sales_earnings = float(input("Enter Sales Earnings: "))
performance_level = float(input("Enter the performance level for this month: "))

# Determine the commission rate based on the performance level
if sales_earnings >= performance_level:
    commission_rate = 0.10
else:
    commission_rate = 0.05

# Calculate the commission
commission = sales_earnings * commission_rate

# Calculate the total salary
total_salary = fixed_salary + commission

# Output the total salary
print("Employee\'s total salary: ", total_salary, "$")
```
