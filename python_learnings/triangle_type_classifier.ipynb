#Problem:

A triangle can be classified as equilateral, isosceles, or scalene based on the lengths of its sides. An equilateral triangle has all three sides of equal length. An isosceles triangle has two sides of equal length and a third side of a different length. If all side lengths are different, the triangle is scalene. Write a program that reads the lengths of the three sides of a triangle from the user. Display a message indicating the type of the triangle.

```python
# Take user input for the lengths of the three sides of a triangle in centimeters.
side_1 = int(input("Enter the length of the first side in cm: "))
side_2 = int(input("Enter the length of the second side in cm: "))
side_3 = int(input("Enter the length of the third side in cm: "))

# Check if the triangle is equilateral (all sides are equal).
if side_1 == side_2 == side_3:
    print("This triangle is an equilateral triangle.")
# Check if the triangle is isosceles (two sides are equal).
elif side_1 == side_2 or side_1 == side_3 or side_2 == side_3:
    print("This triangle is an isosceles triangle.")
# If none of the above conditions are met, the triangle is scalene (all sides are different).
else:
    print("This triangle is a scalene triangle.")
```
