# Description:
This Python code defines a function 'palindrome' to check if a given word is a palindrome.
The program prompts the user to enter a word, converts it to lowercase, removes spaces, and checks if it reads the same backward.
Feedback is provided based on whether the entered word is a palindrome or not.


```python
# Define palindrome fuction
def palindrome(word):
    # Convert the word to lowercase and remove spaces
    word = word.lower().replace(" ", "")

    # Check if the word is equal to its reverse
    result = word == word[::-1]

    return result

# Continuously prompt the user to enter a word and check if it is a palindrome
while True:
    input_word = input("Enter a word: ")
    result = palindrome(input_word)

    # Provide feedback based on whether the word is a palindrome or not
    if result:
        print("The entered '{}' is a palindrome.".format(input_word))
        break
    else:
        print("The entered '{}' is not a palindrome.".format(input_word))
        continue
```
