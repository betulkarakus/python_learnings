
# Description:

In this program, the find_gcd function calculates the Greatest Common Divisor (GCD) using the Euclidean algorithm, and then the find_lcm function utilizes the GCD to find the LCM. The LCM is calculated as the product of the two numbers divided by their GCD. The if __name__ == "__main__": block prompts the user to enter two numbers and then calculates and prints the LCM of those two numbers.

Here's a Python program to find the Least Common Multiple (LCM) of two numbers:

```python
def find_gcd(x, y):
    while y:
        x, y = y, x % y
    return x

def find_lcm(x, y):
    gcd = find_gcd(x, y)
    lcm = (x * y) // gcd
    return lcm

if __name__ == "__main__":
    # Get user input for two numbers
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))

    # Calculate and display the LCM
    lcm_result = find_lcm(num1, num2)
    print(f"The Least Common Multiple (LCM) of {num1} and {num2} is: {lcm_result}")
```
