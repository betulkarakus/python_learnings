
# Description:
This Python script manages a party invitation list. It initially invites the first 3 guests and then prompts the user to add more guests until they decide not to add more. Finally, it displays the total number of attendees and the sorted list of names.


```python
# Create an empty list for the invitation list
name_list = []

# Invite the first 3 guests
for i in range(3):
    invited_guest = input('Enter the name of the guest to invite to the party: ')
    name_list.append(invited_guest)
print("Invited Guests: ", name_list)

# Ask if more guests should be invited
add_more = input('Should more people be invited to the party? (yes/no): ').lower()
while add_more == 'yes':
    # Get the name of the guest to invite
    invited_guest = input('Enter the name of the guest to invite to the party: ')
    # Check if the guest is not already in the list
    if invited_guest not in name_list:
        name_list.append(invited_guest)
    else:
        print(invited_guest, " has already been added to the list...")

    # Ask if more guests should be invited
    add_more = input('Should more people be invited to the party? (yes/no): ').lower()

# Display the total number of attendees and the sorted list of names
print("Number of attendees: ", len(name_list))
name_list.sort(reverse=True)
print("Sorted List of Names: ", name_list)
```
