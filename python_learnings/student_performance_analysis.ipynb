
# Description:
This Python script analyzes the performance of students based on their final grades. It categorizes students into those who passed and those who failed, and prints the results.


```python
# Given final grades and corresponding student names
final_grades = [66, 40, 77, 76, 84, 50, 91]
students = ["Ali", "Kerim", "Jessa", "Emma", "Noah", "Azra", "Kasım"]
passing_students = []
failing_students = []

# Categorize students based on their final grades
for grade in final_grades:
    index = final_grades.index(grade)
    student_name = students[index]
    if grade >= 70:
        passing_students.append(student_name)
    else:
        failing_students.append(student_name)

# Print the lists of passing and failing students
print("Passing Students: ", passing_students)
print("Failing Students: ", failing_students)
```
