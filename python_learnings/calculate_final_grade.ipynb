```python
# Input section
midterm = int(input('Enter your midterm grade: '))
final = int(input('Enter your final exam grade: '))
home_study = int(input('Enter your home study grade: '))
assignment = int(input('Enter your assignment grade: '))
lab = int(input('Enter your lab grade: '))
activity_class = int(input('Enter your activity class grade: '))

# Formula calculation
# The final grade is calculated based on weighted averages of different components.
# Midterm and final exams contribute 25% each, home study contributes 10%, assignment contributes 15%,
# lab contributes 15%, and activity class contributes 10% to the final grade.
final_grade = midterm * 0.25 + final * 0.25 + home_study * 0.10 + assignment * 0.15 + lab * 0.15 + activity_class * 0.10

# Display the result
print('Final Grade: ', final_grade)
```
