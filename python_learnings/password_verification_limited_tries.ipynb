# Description:
This program prompts the user to enter a password and allows a maximum of 3 attempts for correct entry.
If the user enters an incorrect password three times, the program terminates.
Upon successful entry within the allowed attempts, a congratulatory message is displayed; otherwise, access is denied.*italik metin*

```python
#Password Verification Program with Limited Tries

# Define the password
password = "123.qaz"

# Set the maximum number of attempts and initialize the user's attempt count to 0
max_attempt_count = 3
user_attempt_count = 0

# Initialize the permission variable to True
access_permission = True

# Get the password from the user and increment the attempt count
user_password = input('Enter your password: ')
user_attempt_count += 1

# Check if the entered password is correct, repeat until correct or max attempts reached
while user_password != password:
    print("Incorrect password.")

    # Increment the user's attempt count and check if it exceeds the maximum allowed attempts
    user_attempt_count += 1
    if user_attempt_count > max_attempt_count:
        access_permission = False
        break
    else:
        # Get the password again from the user
        user_password = input('Please re-enter your password to verify: ')

# Check if the user successfully entered the password within the allowed attempts
if access_permission:
    print("Congratulations, you have successfully logged in!")
else:
    print("You have reached the maximum number of attempts.")
    print("Access denied!")
```
