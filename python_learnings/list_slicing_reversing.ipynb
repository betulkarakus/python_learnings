```python
# Given list of final exam results
final_results = [90, 92, 83, 45, 56, 87, 98, 78, 88, 10]

# Print the first three elements of the list
print("First three elements:", final_results[:3])

# Print the last three elements of the list
print("Last three elements:", final_results[-3:])

# Print all elements except the last three
print("All elements except last three:", final_results[:-3])

# Print elements starting from the third position to the end
print("Elements from index 2 to the end:", final_results[2:])

# Print the list in reverse order
print("Reversed list:", final_results[::-1])
```
