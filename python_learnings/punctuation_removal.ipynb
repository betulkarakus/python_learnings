
# Description:
This Python program removes punctuation from a given string.


```python
import string

def remove_punctuation(input_string):
    """
    Removes punctuation from a given string.

    Parameters:
    - input_string (str): The input string containing punctuation.

    Returns:
    - str: The input string with punctuation removed.
    """
    translator = str.maketrans("", "", string.punctuation)
    cleaned_string = input_string.translate(translator)
    return cleaned_string

if __name__ == "__main__":
    # Get user input for the string
    input_string = input("Enter a string with punctuation: ")

    # Remove punctuation from the string
    result_string = remove_punctuation(input_string)

    # Display the result
    print("\nString without punctuation:")
    print(result_string)
```
