```python
# Given shopping list
shopping_list = ["Nutella", "Peanuts", "Tea", "Chocolate", "Coffee"]
print("Original shopping list:", shopping_list)

# Add a new item to the shopping list
shopping_list.append("Yogurt")
print("Shopping list after adding a new item:", shopping_list)

# Insert a new item between two existing items in the list
shopping_list.insert(2, "Napkin")
print("Shopping list after inserting a new item:", shopping_list)

# Add two new items to the end of the list
shopping_list.extend(["Biscuit", "Cleaning Agent", "Lettuce"])
print("Shopping list after adding two items to the end:", shopping_list)

# Replace two consecutive items in the list with new items
shopping_list[1:3] = ["Parsley", "Cola"]
print("Shopping list after replacing two items:", shopping_list)

# Convert all items in the list to lowercase
for index in range(len(shopping_list)):
    shopping_list[index] = shopping_list[index].lower()

print("Shopping list with all items in lowercase:", shopping_list)
```
