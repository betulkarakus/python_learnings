
# Description:
This Python program demonstrates two methods for merging two dictionaries.



```python
# Method 1: Using the update() method
def merge_dicts_method1(dict1, dict2):
    """
    Merge two dictionaries using the update() method.

    Parameters:
    - dict1: First dictionary to be merged.
    - dict2: Second dictionary to be merged.

    Returns:
    - Merged dictionary.
    """
    merged_dict = dict1.copy()
    merged_dict.update(dict2)
    return merged_dict

# Method 2: Using the {**dict1, **dict2} syntax
def merge_dicts_method2(dict1, dict2):
    """
    Merge two dictionaries using the {**dict1, **dict2} syntax.

    Parameters:
    - dict1: First dictionary to be merged.
    - dict2: Second dictionary to be merged.

    Returns:
    - Merged dictionary.
    """
    merged_dict = {**dict1, **dict2}
    return merged_dict

# Example dictionaries
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}

# Method 1
merged_dict_method1 = merge_dicts_method1(dict1, dict2)
print("Merged Dictionary (Method 1):", merged_dict_method1)

# Method 2
merged_dict_method2 = merge_dicts_method2(dict1, dict2)
print("Merged Dictionary (Method 2):", merged_dict_method2)
```
