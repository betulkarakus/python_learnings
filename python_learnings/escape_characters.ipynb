# Description:
If you use double quotes within a string enclosed in double quotes, you will encounter an error.


```python
# Incorrect string with double quotes
txt = "We are the "Vikings of the North"."
```

The escape character allows you to use double quotes within a string.

```python
txt = "We are the North \"Vikings\"."
print(txt)
```

Common escape characters used in Python:

- \'	Single Quote
- \\	Backslash	  
- \n	New Line	    
- \r	Carriage Return
- \t	Tab	         
- \b	Backspace
