```python
# Define variables & take user inputs.
monthly_internet_usage_hours = int(input("Enter your monthly total internet usage hours: "))
category = int(input("Enter your internet category number: "))

# Check if the entered category is valid (between 1 and 3).
if 1 <= category <= 3:
    monthly_bill = 0  # Initially set the monthly bill to zero.

    # Check the internet usage range and calculate the monthly bill accordingly.
    if 0 <= monthly_internet_usage_hours < 90:
        if category == 1:
            monthly_bill = 1300
        elif category == 2:
            monthly_bill = 1500
        elif category == 3:
            monthly_bill = 1700
    elif 90 <= monthly_internet_usage_hours < 150:
        if category == 1:
            monthly_bill = 1200
        elif category == 2:
            monthly_bill = 1400
        elif category == 3:
            monthly_bill = 1600
    elif 150 <= monthly_internet_usage_hours < 300:
        if category == 1:
            monthly_bill = 1100
        elif category == 2:
            monthly_bill = 1300
        elif category == 3:
            monthly_bill = 1500
    elif 300 <= monthly_internet_usage_hours:
        if category == 1:
            monthly_bill = 1000
        elif category == 2:
            monthly_bill = 1200
        elif category == 3:
            monthly_bill = 1400

    # Check if the monthly bill is still zero (invalid internet usage hours).
    if monthly_bill == 0:
        print("Invalid internet usage hours!")
    else:
        print("Your total monthly bill is", monthly_bill, "$.")
else:
    print("You entered an invalid category! Please check your category and try again.")

print("End of the program")
```
