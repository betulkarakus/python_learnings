
# Description:
This Python program checks if a given number is a Happy Number.


A Happy Number is a number that eventually reaches 1 when replaced by the sum of the square of each digit. If a number leads to an endless loop in this process without ever reaching 1, it is not a Happy Number.

For example, 19 is a happy number because the process ends with 1:  

1^2 + 9^2 = 82,

8^2 + 2^2 = 68,

6^2 + 8^2 = 100,

1^2 + 0^2 + 0^2 = 1.



```python
def is_happy_number(number):
    """
    Checks if the given number is a Happy Number.

    Parameters:
    - number (int): The input number to be checked.

    Returns:
    - bool: True if the number is a Happy Number, False otherwise.
    """
    def get_next_number(n):
        return sum(int(digit) ** 2 for digit in str(n))

    seen_numbers = set()

    while number != 1 and number not in seen_numbers:
        seen_numbers.add(number)
        number = get_next_number(number)

    return number == 1

if __name__ == "__main__":
    # Get user input for the number
    input_number = int(input("Enter a number: "))

    # Check if the number is a Happy Number
    result = is_happy_number(input_number)

    # Display the result
    if result:
        print(f"{input_number} is a Happy Number.")
    else:
        print(f"{input_number} is not a Happy Number.")
```
