```python
# Input section
# Constants: Universal gas constant (R)
R = 0.08206

# Get user inputs for the number of moles (n), temperature (T), and volume (V)
n = float(input('Enter the number of moles: '))
T = float(input('Enter the temperature: '))
V = float(input('Enter the volume: '))

# Formula calculation
# Calculate the pressure (P) using the Ideal Gas Law formula: P = (nRT) / V
# where P is pressure, n is the number of moles, R is the universal gas constant, T is temperature, and V is volume.
P = (n * R * T) / V

# Display the result
print('Pressure: ', P)
```
