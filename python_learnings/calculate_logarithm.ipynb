
# Description:
This program takes a positive number as input, calculates the natural logarithm using the **math.log()** function, and then prints the result. It also includes a check to ensure that the input number is positive, as the natural logarithm is undefined for non-positive numbers.


```python
import math

# Function to calculate natural logarithm
def calculate_logarithm(number):
    if number > 0:
        result = math.log(number)
        return result
    else:
        print("Please enter a positive number.")

# Get user input for the number
user_number = float(input("Enter a positive number to calculate its natural logarithm: "))

# Calculate natural logarithm using the function
log_result = calculate_logarithm(user_number)

# Display the result
if log_result is not None:
    print(f"The natural logarithm of {user_number} is: {log_result:.4f}")
```
