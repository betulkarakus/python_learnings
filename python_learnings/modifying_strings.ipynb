# Description:
Python provides built-in methods for manipulating strings.


```python
# Original string
a = "Hello World!"

# Convert the string to uppercase
a_uppercase = a.upper()

# Print the result
print(a_uppercase)
```

```python
# Original string
a = "Hello World!"

# Convert the string to lowercase
a_lowercase = a.lower()

# Print the result
print(a_lowercase)
```

```python
# Original string with extra spaces
a = "   Hello World!   "

# Strip leading and trailing whitespace
b = a.strip()

# Print the original and stripped strings
print(a)
print(b)
```

```python
# Original string
a = "Hello World!"

# Replace 'o' with 'ö'
b = a.replace("o", "ö")

# Print the modified and original strings
print(b)
print(a)
```

```python
# Original string
a = "Hello, World!"

# Split the string at the comma
split_result = a.split(",")

# Print the resulting substrings as a list
print(split_result)
```
