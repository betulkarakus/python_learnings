```python
# A program to calculate the total daily sales

# Initialize the total sales to zero
total_sales = 0

# Get the sales amount from the user
sales = int(input("Enter the sales amount, enter -1 to exit: "))

# Continue getting sales amounts until -1 is entered
while sales != -1:
    # Add the current sales amount to the total
    total_sales += sales
    # Get the next sales amount
    sales = int(input("Enter the sales amount: "))

# Print the total sales for the day
print("Total sales for today: ", total_sales)
```
