```python
# Sum of all ODD numbers from N to M
# Get the lower limit (N) from the user
lower_limit = int(input("Enter the value of N: "))
# Get the upper limit (M) from the user
upper_limit = int(input("Enter the value of M: "))

# Initialize the sum of odd numbers to zero
sum_odd_numbers = 0

# Iterate through the range from N to M
while lower_limit < upper_limit:
    # Check if the current number is odd
    if lower_limit % 2 == 1:
        # If it's odd, add it to the sum
        sum_odd_numbers += lower_limit
        # Move to the next odd number (skip even numbers)
        lower_limit += 2
    else:
        # If the current number is even, move to the next number
        lower_limit += 1

# Print the total sum of odd numbers in the specified range
print("Total: ", sum_odd_numbers)
```
