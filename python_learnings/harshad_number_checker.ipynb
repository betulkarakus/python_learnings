
# Description:
This Python program checks if a given number is a Harshad Number.

A Harshad Number (also called Niven number) is a number that is divisible by the sum of its digits. In other words, a number is a Harshad Number if it is divisible by the sum of its digits without any remainder.

For example:
- 18 is a Harshad number because , and 18 is divisible by 9.
- 42 is not a Harshad number because , and 42 is not divisible by 6.



```python
def is_harshad_number(number):
    """
    Checks if the given number is a Harshad Number.

    Parameters:
    - number (int): The input number to be checked.

    Returns:
    - bool: True if the number is a Harshad Number, False otherwise.
    """
    # Calculate the sum of digits
    digit_sum = sum(int(digit) for digit in str(number))

    # Check if the number is divisible by the sum of its digits
    return number % digit_sum == 0

if __name__ == "__main__":
    # Get user input for the number
    input_number = int(input("Enter a number: "))

    # Check if the number is a Harshad Number
    result = is_harshad_number(input_number)

    # Display the result
    if result:
        print(f"{input_number} is a Harshad Number.")
    else:
        print(f"{input_number} is not a Harshad Number.")
```
