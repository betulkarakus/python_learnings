
# Description:
This program defines a function to calculate the sum of two numbers and utilizes it to add two numbers entered by the user. The result is then displayed with a formatted message.


```python
# Function definition to calculate the sum of two numbers
def calculate_sum(a, b):
    total = a + b
    return total

# Get the first number from the user
number_1 = int(input("Enter the first number you want to add: "))

# Get the second number from the user
number_2 = int(input("Enter the second number you want to add: "))

# Call the function with the user-input numbers
result = calculate_sum(number_1, number_2)

# Print the result
print("The sum of {} and {} is {}.".format(number_1, number_2, result))
```
