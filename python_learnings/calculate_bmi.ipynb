
# Description:
This program takes the user's weight in kilograms and height in meters as input, calculates the BMI using the provided formula, and then prints the BMI value along with an interpretation of the result based on standard BMI categories.

A simple Python program to calculate Body Mass Index (BMI):


```python
# Python program to calculate Body Mass Index (BMI)

# Function to calculate BMI
def calculate_bmi(weight_kg, height_m):
    bmi = weight_kg / (height_m ** 2)
    return bmi

# Get user input for weight and height
weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))

# Calculate BMI using the function
bmi_result = calculate_bmi(weight, height)

# Display the BMI result
print(f"Your Body Mass Index (BMI) is: {bmi_result:.2f}")

# Interpretation of BMI categories
if bmi_result < 18.5:
    print("You are underweight.")
elif 18.5 <= bmi_result < 25:
    print("Your weight is normal (healthy range).")
elif 25 <= bmi_result < 30:
    print("You are overweight.")
else:
    print("You are obese.")
```
