
# Description:
This Python code prompts the user to enter a password and checks its strength based on specific criteria.
The criteria include a minimum length of 8 characters, at least one uppercase letter, one lowercase letter, and one digit.
The code then provides feedback on the strength of the password.


```python
# Prompt the user to enter a password
password_input = input("Please enter your password: ")

# Define the function to check password strength
def password_check(password):
    length_condition = False
    uppercase_condition = False
    lowercase_condition = False
    digit_condition = False

    # Check if the password meets the length criterion (at least 8 characters)
    if len(password) >= 8:
        length_condition = True

    # Iterate through each character in the password to check for uppercase, lowercase, and digit conditions
    for character in password:
        if 65 <= ord(character) <= 90:  # Check for uppercase
            uppercase_condition = True
        elif 97 <= ord(character) <= 122:  # Check for lowercase
            lowercase_condition = True
        elif 48 <= ord(character) <= 57:  # Check for digit
            digit_condition = True

    return length_condition, uppercase_condition, lowercase_condition, digit_condition

# Call the password_check function and store the results
length_condition, uppercase_condition, lowercase_condition, digit_condition = password_check(password_input)

# Determine the overall strength of the password
if length_condition and uppercase_condition and lowercase_condition and digit_condition:
    result = "STRONG"
else:
    result = "WEAK"

    # Determine which specific criterion is not met
    if not length_condition:
        unmet_criterion = "length of at least 8 characters"
    elif not uppercase_condition:
        unmet_criterion = "at least one uppercase letter"
    elif not lowercase_condition:
        unmet_criterion = "at least one lowercase letter"
    elif not digit_condition:
        unmet_criterion = "at least one digit"

# Provide feedback based on password strength
if result == "STRONG":
    print("Your password is STRONG as it meets all the criteria.")
else:
    print("Your password is considered WEAK because it does not meet the '{}' criterion.".format(unmet_criterion))
```
