
# Description:
This program uses the ord() function, which returns an integer representing the Unicode character. In the case of ASCII, Unicode and ASCII values are the same for characters in the ASCII character set.

A simple Python program that takes a character as input and prints its ASCII value:


```python
# Python program to find the ASCII value of a character

# Get a character input from the user
char_input = input("Enter a character: ")

# Check if the input is a single character
if len(char_input) == 1:
    # Use the ord() function to get the ASCII value
    ascii_value = ord(char_input)

    # Print the result
    print(f"The ASCII value of '{char_input}' is {ascii_value}")
else:
    print("Please enter only one character.")
```
