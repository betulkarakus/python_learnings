
# Description:
This Python script takes a list of numbers and creates a new list, where 50 is added to each element that is less than or equal to 100 in the original list. The resulting list is then printed.

```python
# Given list of numbers
number_list = [200, 7, 90, 800, 80]
new_number_list = []

# Create a new list by adding 50 to elements in number_list that are less than or equal to 100
for element in number_list:
    if element <= 100:
        element += 50
    new_number_list.append(element)

# Print the new list
print(new_number_list)
```
