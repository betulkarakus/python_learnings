# Scenario 1:
If the student is successful in all courses, they can move to the next grade.



```python
# Grades for each subject
final_grades = [70, 70, 70, 76, 84, 80, 91]
pass_fail_status = []

# Check if the student passed or failed each course
for grade in final_grades:
    if grade >= 70:
        pass_fail_status.append(True)
    else:
        pass_fail_status.append(False)

# Check if the student passed all courses to determine promotion eligibility
if all(pass_fail_status):
    print("The student has successfully advanced to the next grade.")
else:
    print("The student could not advance to the next grade.")
```


# Scenario 2:
If the student has only one failing grade, they can still move to the next grade.

```python
# Grades for each subject
final_grades = [66, 70, 77, 76, 84, 80, 91]
pass_fail_status = []

# Check if the student passed or failed each course
for grade in final_grades:
    if grade >= 70:
        pass_fail_status.append(True)
    else:
        pass_fail_status.append(False)

# Check if the student has only one failing grade to determine promotion eligibility
if pass_fail_status.count(False) < 2:
    print("The student has successfully advanced to the next grade.")
else:
    print("The student could not advance to the next grade.")
```

# Scenario 3:
If the student passes at least one course, they can move to the next grade.

```python
# Grades for each subject
final_grades = [40, 40, 40, 76, 60, 30, 20]
pass_fail_status = []

# Check if the student passed or failed each course
for grade in final_grades:
    if grade >= 70:
        pass_fail_status.append(True)
    else:
        pass_fail_status.append(False)

# Check if the student passed at least one course to determine promotion eligibility
if pass_fail_status.count(True) > 0:
    print("The student has successfully advanced to the next grade.")
else:
    print("The student could not advance to the next grade.")
```
