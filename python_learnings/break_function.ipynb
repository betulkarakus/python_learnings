```python
# Get the control value from the user
control_value = int(input("Enter the control value: "))

# Iterate through numbers from 1 to 10
for number in range(1, 11):
    # Check if the current number is equal to the control value
    if number == control_value:
        # If equal, print a message and exit the loop using 'break'
        print("Exiting the loop...")
        break
        # The following statement is never executed because of the 'break' above
        print("This statement is never executed")

    # Print the current number within the loop
    print("Number: ", number)

# Print the first statement after the loop
print("First statement after the loop")
```
