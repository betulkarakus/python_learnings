
# Description:
This Python program defines a base class 'Person' and two child classes 'Male' and 'Female'.
Each class has a method 'getGender' that prints the gender of the person.



```python
class Person:
    def getGender(self):
        """
        Print the gender of the person.
        """
        print("Undefined Gender")

class Male(Person):
    def getGender(self):
        """
        Print "Male" as the gender.
        """
        print("Male")

class Female(Person):
    def getGender(self):
        """
        Print "Female" as the gender.
        """
        print("Female")

# Example usage
person = Person()
male_person = Male()
female_person = Female()

# Calling getGender for each instance
person.getGender()      # Output: Undefined Gender
male_person.getGender()  # Output: Male
female_person.getGender()  # Output: Female
```
