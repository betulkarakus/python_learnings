
# Description:
This program allows the user to test whether a given number is odd or even.
After each test, the user is prompted to continue the process or exit.
The program ensures the user enters only valid responses ('y' or 'n') for continuation.


```python
# Function definition
def odd_even(number):
    # Test if the number is even
    if number % 2 == 0:
        result = 0  # 0 represents even
    else:
        result = 1  # 1 represents odd

    # Return the content of the 'result' variable to the point where the function is called
    return result

# Function for continuing the process control
def continue_processing():
    # Ask the user for their preference
    response = input("Do you want to continue the process? (y/n): ").lower()

    # Check if the user pressed the correct key.
    # Create a loop until the correct letters "y" or "n" are pressed.
    while True:
        if response == "n" or response == "y":
            # The user pressed the correct key.
            # Exit this loop.
            break
        else:
            # The user pressed the wrong key.
            print("Please press either 'y' or 'n'!")
            response = input("Do you want to continue the process? (y/n): ").lower()

    # Set the return value of the function based on the user's response.
    if response == "n":
        return False
    else:
        return True

# Main program

# Create an infinite loop
while True:
    # Get the number for the test from the user
    num_of_tests = int(input("Enter the number to test: "))

    # Call the function with its name and argument
    result_odd_even = odd_even(num_of_tests)

    # Print the result
    if result_odd_even == 0:
        print(num_of_tests, "is an EVEN number.")
    else:
        print(num_of_tests, "is an ODD number.")

    # Should the test process continue?
    continue_process = continue_processing()
    if continue_process == False:
        break

print("Have a nice day!")
```
